```python
import pandas as pd 
import os
```

```python
diretorio = '/home/data-ramiro/Documents/bootcamp_pan/BASE_TP/'
```

```python
lista_arquivos = os.listdir(diretorio)
```

```python
lista_arquivos
```

```python
lista_df = []
for arquivo in lista_arquivos:
    caminho_completo = os.path.join(diretorio, arquivo)
    df = pd.read_csv(caminho_completo, sep="|", encoding='latin-1')
    lista_df.append(df)
```

```python
df_total = pd.concat(lista_df)
df_total
```

```python
df_pesquisa = df_total
df_pessoa = df_total[['cod_pessoa', 'genero', 'data_nascimento']].drop_duplicates()
df_animal_estimacao = df_total[['animal_estimacao']].drop_duplicates()
df_clima = df_total[['clima']].drop_duplicates()
df_bebida_favorita = df_total[['bebida_favorita']].drop_duplicates()
df_hobbies = df_total[['hobbies']].drop_duplicates()
```

```python
import MySQLdb
import sqlalchemy
```

```python
user = 'root'
password = '******'
host = '127.0.0.1'
database = 'pesquisa'

string_conexao = f'mysql://{user}:{password}@{host}/{database}'
string_conexao
```

```python
engine = sqlalchemy.create_engine(string_conexao)
conn = engine.connect()
```

```python
df_pessoa.itertuples()
```

```python
for pessoa in df_pessoa.itertuples():
    codP = pessoa.cod_pessoa
    gen = pessoa.genero
    dtn = pessoa.data_nascimento
    
    try:
        sql = f"""INSERT INTO pessoa (cod_pessoa, genero, data_nascimento) VALUES
                                     ('{codP}', '{gen}', '{dtn}')"""
        
        #print(sql)
        
        conn.execute(sql)
        
    except Exception as e:
        print(f'Não foi possível inserir o registro {codP}. O erro foi: {e}')
    
```

```python
df_animal_estimacao = pd.read_sql('animal_estimacao', conn)
df_clima = pd.read_sql('clima', conn)
df_bebida_favorita = pd.read_sql('bebida', conn)
df_hobbies = pd.read_sql('hobbie', conn)
#df_animal_estimacao
#df_clima
#df_bebida_favorita
#df_hobbies
```

```python
dict_a = df_animal_estimacao.set_index(df_animal_estimacao.animal)['cod_animal_estimacao'].to_dict()
dict_c = df_clima.set_index(df_clima.clima)['cod_clima'].to_dict()
dict_b = df_bebida_favorita.set_index(df_bebida_favorita.bebida)['cod_bebida'].to_dict()
dict_h = df_hobbies.set_index(df_hobbies.hobbie)['cod_hobbie'].to_dict()
#dict_a
#dict_c
#dict_b
#dict_h
```

```python
df_pesquisa.animal_estimacao = df_pesquisa.animal_estimacao.map(dict_a)
df_pesquisa.clima = df_pesquisa.clima.map(dict_c)
df_pesquisa.bebida_favorita = df_pesquisa.bebida_favorita.map(dict_b)
df_pesquisa.hobbies = df_pesquisa.hobbies.map(dict_h)
df_pesquisa
```

```python
for pesquisa in df_pesquisa.itertuples():
    codP = pesquisa.cod_pessoa
    dtcol = pesquisa.data_coleta
    codA = pesquisa.animal_estimacao
    codC = pesquisa.clima
    codB = pesquisa.bebida_favorita
    codH = pesquisa.hobbies
    
    try:
        sql = f"""INSERT INTO pesquisa  (data_coleta, cod_pessoa, cod_animal_estimacao, cod_bebida, cod_hobbie, cod_clima) VALUES
                                        ('{dtcol}', '{codP}', '{codA}', '{codB}', '{codH}', '{codC}')"""
        
        #print(sql)
        
        conn.execute(sql)
        
    except Exception as e:
        print(f'Não foi possível inserir o registro {dtcol}. O erro foi: {e}')
    
```
